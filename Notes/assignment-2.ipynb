```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/pakistan-crimes-by-type/248a92b9-5817-43f2-b0e7-6d45a45a8ee9.csv')
df
```

```python
df['Year'].unique()
```

***Robberies in Pakistan 2012 to 2017***

```python
robbery = df[df['Offence'] == 'Robbery']
robbery = robbery[['Year', 'Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways']]
robbery = robbery.melt(id_vars=['Year'], var_name='Province', value_name='Robbery Cases')
robbery = robbery.sort_values(by='Robbery Cases', ascending=False)
robbery
```

```python
totalCasesOfRobberies = df[df['Offence'] == 'Robbery'].agg({'Punjab': 'sum', 'Sindh': 'sum', 'KP': 'sum', 'Balochistan': 'sum', 'Islamabad': 'sum', 'Railways': 'sum', 'G.B': 'sum', 'AJK': 'sum', 'Pakistan': 'sum', }).reset_index()
totalCasesOfRobberies.columns = ['Province', 'Total Robbery Cases']
totalCasesOfRobberies = totalCasesOfRobberies.sort_values(by='Total Robbery Cases', ascending=False)
totalCasesOfRobberies
```

```python
import pandas as pd
import numpy as np
```

```python
# Create a sample DataFrame
data = {'A': [1, 2, np.nan, 4],
        'B': [np.nan, 2, 3, 4],
        'C': [1, np.nan, np.nan, 4]}

df = pd.DataFrame(data)
df
```

```python
# Identify missing values

df.isnull().sum()
#print(df.isnull().sum())
```

```python
# Drop rows with any missing values

df_dropped = df.dropna()
df_dropped
```

```python
# Fill missing values with a specific value

df_filled = df.fillna(0)
df_filled
```

```python
#forward fill missing values

df_ffill=df.fillna(method='ffill')
df_ffill
```

```python
#fill with the mean

df_mean=df.copy()
df_mean['A']=df_mean['A'].fillna(df_mean['A'].mean())
df_mean['B']=df_mean['B'].fillna(df_mean['B'].mean())
df_mean['C']=df_mean['C'].fillna(df_mean['C'].mean())
df_mean
```

```python
#Interpolation

df_interpolation=df.interpolate()
df_interpolation
```

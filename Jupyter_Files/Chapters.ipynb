Arithmetic Operations

```python
var=pd.DataFrame({
    'A':[1,2,3,4],
    'B':[5,6,7,8]
})
```

```python
var['C']=var['A'] - var['B']
var
```

```python
var1=pd.DataFrame({
    'A':[10,20,30,40],
    'B':[55,66,77,88]
})
var1
```

```python
var1['python']=var1['A'] <= 20
var1['python_1']=var1['B'] >= 60
var1
```

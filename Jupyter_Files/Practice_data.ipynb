```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\progr\\OneDrive\\Desktop\\DS\\Files\\mckinsey_newdata.csv')
df
```

```python
del df['Unnamed: 0']
```

```python
df['Country_Names'].unique()
```

```python
pakistan_data=df[df['Country_Names'] == 'Pakistan']
India_data=df[df['Country_Names'] == 'India']
China_data=df[df['Country_Names'] ==  'China']
pakistan_data
```

```python
Asia_data=pd.concat([pakistan_data,India_data,China_data], ignore_index=True)
Asia_data.head()
```

```python
plt.bar('year','GDP_Capita',data=pakistan_data ,color='green',edgecolor='black')
plt.title('Pakistan GDP per capita',color='b',fontweight='bold', fontsize=10)
plt.xlabel('Years', fontweight='bold', fontsize=10)
plt.ylabel('GDP per capita',fontweight='bold', fontsize=10)
plt.show()
```

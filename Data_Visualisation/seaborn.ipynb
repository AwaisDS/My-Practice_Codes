```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data={
    'A':[1,2,3,4,5,6],
    'B':[3,4,5,6,7,8]
}
df=pd.DataFrame(data)
df
```

***Line Plot***

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='A' , y='B' , data=df , color='red')
plt.show()
```

```python
penguins_data=sns.load_dataset('penguins').head(20)
penguins_data
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data )
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data , hue='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data , hue='sex' , size=10)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data , hue='sex' , size=10 , style='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data , hue='sex' , size=10 , style='sex' , palette='Accent_r')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , data=penguins_data , hue='sex' , size=10 , style='sex' , palette='Accent_r' , markers=['o' , '>'])
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , 
                data=penguins_data ,
                hue='sex' , size=10 , 
                style='sex' , 
                palette='Accent_r' , 
                markers=['o' , '>'] ,
                dashes=False
                )
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , 
                data=penguins_data ,
                hue='sex' , size=10 , 
                style='sex' , 
                palette='Accent_r' , 
                markers=['o' , '>'] ,
                dashes=False , 
                legend='brief'
                )
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='bill_depth_mm' , 
                data=penguins_data ,
                hue='sex' , size=10 , 
                style='sex' , 
                palette='Accent_r' , 
                markers=['o' , '>'] ,
                dashes=False , 
                legend='brief' , 
                )
plt.grid()
plt.title('This is Title')
plt.show()
```

***Histogram***

```python
penguins_data_2=sns.load_dataset('penguins')
penguins_data_2.head()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'])
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , bins=[170,180,190,200,210,220,230] )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            rug=True , 
            )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            rug=True , 
            color='g')
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            kde=True , 
            rug=True , 
            color='g',
            log_scale=True)
plt.show()
```

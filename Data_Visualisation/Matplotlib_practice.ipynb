```python
# set width of bar 
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

# set height of bar 
IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 

# Set position of bar on X axis 
br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth, 
        edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, 
        edgecolor ='grey', label ='ECE') 
plt.bar(br3, CSE, color ='b', width = barWidth, 
        edgecolor ='grey', label ='CSE') 

# Adding Xticks 
plt.xlabel('Branch', fontweight ='bold', fontsize = 15) 
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], 
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show() 
```

```python
# Data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4','Category 5','Category 6','Category 7']
values_option1 = [5, 7, 8, 6, 13, 10, 16]
values_option2 = [14, 6, 7, 5, 7, 8, 18]
values_option3 = [3, 18, 5, 9, 21, 20, 23]
values_option4= [4, 6, 19, 3, 11, 15, 25]

# Number of categories
n = len(categories)

# Bar width
bar_width = 0.20
fig = plt.subplots(figsize =(10, 8))
# Set position of bar on X axis
r1 = np.arange(n)
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Create grouped bar plot
plt.bar(r1, values_option1, color='b', width=bar_width, edgecolor='black', label='Option 1')
plt.bar(r2, values_option2, color='g', width=bar_width, edgecolor='black', label='Option 2')
plt.bar(r3, values_option3, color='r', width=bar_width, edgecolor='black', label='Option 3')
plt.bar(r4, values_option4, color='y', width=bar_width, edgecolor='black', label='option 4')
# Add labels and title
plt.xlabel('Category', fontweight='bold')
plt.ylabel('Values', fontweight='bold')
plt.title('Grouped Bar Plot with Three Options')
plt.xticks([r + bar_width for r in range(n)], categories)

# Add legend
plt.legend()

# Display the plot
plt.show()
```

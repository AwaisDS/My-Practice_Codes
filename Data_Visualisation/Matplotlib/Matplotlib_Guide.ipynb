```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
#y=[7,8,9,10,11,12]
```

```python
#basic line plot
y=[4,7,2,9,3,0]
plt.plot(x,y)

plt.title('One line plot')
plt.xlabel('X-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#creating a scatter plot

plt.scatter(x,y)

plt.title('Scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#bar plot

o=['A','B','C','D']
p=[1,2,3,4]

plt.bar(o,p)

plt.title('Simple bar Plot')
plt.xlabel('Catagories')
plt.ylabel('values')

plt.show()
```

```python
# Creating  histogram

data=np.random.randn(1000)

plt.hist(data, bins=30, range=(-3, 3), density=True, alpha=0.7,
         color='blue', edgecolor='black', linewidth=1.2, histtype='bar', rwidth=0.9)

plt.title('Simple Histogram')
plt.xlabel('value')
plt.ylabel('Frequency')

plt.show()
```

```python
#plot with legend

plt.plot(x,y ,label='Line 1')

plt.title('plot with legend')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.show()
```

```python
#Customizing the plot

plt.plot(x,y,label='--',marker='o',color='r')

plt.title('Customized plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#create subplots

plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('plot 1')
plt.xlabel('x axis')

plt.subplot(1,3,2)
plt.bar(x,y,color='red')
plt.title('plot 2')

plt.subplot(1,3,3)
plt.scatter(x,y)
plt.title('plot 3')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 6, 8, 10]
y2 = [2, 2, 7, 10, 12]

plt.fill_between(x, y1, color="skyblue", alpha=0.5)
plt.fill_between(x, y2, color="olive", alpha=0.5)
plt.title("Area Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Box Plot Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python
import seaborn as sns

data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, cmap='viridis')
plt.title("Heatmap Example")
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.violinplot(data, showmedians=True)
plt.title("Violin Plot Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

ax.scatter(x, y, z)
plt.title("3D Plot Example")
plt.show()
```

```python
# Generate random data
data = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'z': np.random.randn(100)
})

sns.pairplot(data)
plt.title("Pair Plot")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('anagrams')
df.head()
```

```python
hm = df.drop(columns='attnr' , axis=1).head(10)
```

```python
sns.heatmap(hm)
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
sns.heatmap(var)
```

```python
sns.heatmap(
    var,
    vmin=0,
    vmax=10
)
```

```python
sns.heatmap(
    var,
    vmin=0,
    vmax=10,
    cmap='viridis'
)
```

```python
sns.heatmap(
    var,
    vmin=0,
    vmax=10,
    cmap='viridis',
    annot=True
    
)
```

```python
sns.heatmap(
    var,
    vmin=0,
    vmax=10,
    cmap='viridis',
    annot=True,
    linewidths=5
)
```

```python
v=sns.heatmap(
    var,
    vmin=0,
    vmax=10,
    cmap='viridis',
    annot=True,
    linewidths=5,
    linecolor='g'
)
v.set(xlabel='X axis' , ylabel='y axis')
plt.show()
```

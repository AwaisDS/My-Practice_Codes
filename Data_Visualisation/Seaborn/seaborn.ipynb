***Histogram***

```python
penguins_data_2=sns.load_dataset('penguins')
penguins_data_2.head()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'])
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , bins=[170,180,190,200,210,220,230] )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            rug=True , 
            )
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            bins=[170,180,190,200,210,220,230] , 
            kde=True , 
            rug=True , 
            color='g')
plt.show()
```

```python
sns.displot(penguins_data_2['flipper_length_mm'] , 
            kde=True , 
            rug=True , 
            color='g',
            log_scale=True)
plt.show()
```

***Bar Plot***

```python
bar_data=sns.load_dataset('penguins')
```

```python
bar_data.head()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            )
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex')
plt.show()
```

```python
order_1=['Dream' , 'Biscoe' , 'Torgersen']
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            order=order_1
            )
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            hue_order=['Female','Male'])
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', False),
            edgecolor='black'
            )
plt.show()
```

```python
sns.barplot(x='bill_depth_mm' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', False),
            orient='h'
            )
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data,
            color='g')
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', False),
            edgecolor='black',
            palette='icefire'
            )
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', False),
            edgecolor='black',
            palette='icefire',
            saturation=0.3
            )
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', 65),
            edgecolor='black',
            palette='icefire',
            errcolor='y'            )
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', 65),
            edgecolor='black',
            palette='icefire',
            errcolor='y',
            capsize=0.3            )
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', 65),
            edgecolor='black',
            palette='icefire',
            errcolor='y',
            dodge=False)
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', 65),
            edgecolor='black',
            palette='icefire',
            errcolor='y',
            alpha=0.3)
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

```python
sns.set(style='dark')
sns.barplot(x='island' ,
            y='bill_length_mm' ,
            data=bar_data ,
            hue='sex',
            errorbar=('ci', 65),
            edgecolor='black',
            palette='icefire',
            errcolor='y')
plt.title('Bill length acc to Island' , fontsize=16)
plt.show()
```

***Scatter Plot***

```python
scatter_data=sns.load_dataset('penguins')
```

```python
scatter_data.head(3)
```

```python
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm' , data=scatter_data)
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.set(style='dark')
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex',
                style='sex')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex',
                style='sex',
                size='sex')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex',
                style='sex',
                size='sex',
                sizes=[100,50])
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex',
                style='sex',
                size='sex',
                sizes=[100,50],
                palette='turbo')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
m={'Male':'*' , 'Female':'o'}
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm' , 
                data=scatter_data ,
                hue='sex',
                style='sex',
                size='sex',
                sizes=[100,50],
                palette='turbo',
                markers=m)
plt.title('Bill Length vs Bill Depth')
plt.show()
```
